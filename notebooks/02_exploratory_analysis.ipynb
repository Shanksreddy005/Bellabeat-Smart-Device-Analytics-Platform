# Phase 4: Analyze (Exploratory Data Analysis)
This notebook explores general distributions, trends, and correlations in the cleaned datasets.

```
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sys.path.append('../scripts')
import config

df = pd.read_csv('../data/processed/cleaned_daily_activity.csv')
print(df.describe())
```

### Correlation Heatmap

```
plt.figure(figsize=(10, 8))
corr = df[['TotalSteps', 'TotalDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'Calories']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Daily Metrics')
plt.show()
```
